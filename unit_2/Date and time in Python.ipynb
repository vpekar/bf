The last two are part of the Anaconda distribution and are already installed on your computer if you have installed Anaconda.

# Time

At the most basic level, time is represented in Python (as in many other programming languages) as the number of seconds since a selected point in time, known as "epoch", which is 12:00am, January 1, 1970.

We can find out how many seconds have passed from the epoch to the current point in time, as follows:

```python
import time

x = time.time()
x
```

Note the returned value is a float, that is, Python measures also fractions of seconds.

The `time.time` function can be used, for example, to determine how long it takes for a program or a certain part of it to run:

```python
# note the time when the program started
start = time.time()

# pause for 3 seconds, for illustration
time.sleep(3)

# note the time the program ended
end = time.time()

# print the difference
x = end - start
print(f"It took {x} seconds.")
```

# Datetime

In many data analysis projects, it is often necessary to manipulate dates. For this purpose, the "datetime" package is commonly used.

One can obtain the current date and time, in [UTC](https://en.wikipedia.org/wiki/Coordinated_Universal_Time) (i.e., the time standard against which all the world’s timekeeping is synchronized):

```python
from datetime import datetime

utcnow = datetime.utcnow()
utcnow
```

The variable `utcnow` is a datetime object. It has attributes that store the year, month, day, hour, minute, second, and microsecond:

```python
print(f"year {utcnow.year}, quarter {utcnow}. month {utcnow.month},",
      f"day {utcnow.day}, hour {utcnow.hour}, minute {utcnow.minute}")
```

To look up the current time on your local machine, do:

```python
now = datetime.now()
now
```

## Manipulating datetime objects

One can create a datetime object for any given date and time:

```python
# object for 20-06-2000, 15:00:00
then = datetime(year=2000, month=6, day=20, hour=15, minute=0, second=0)
then
```

Find the difference between now and the date in the past:

```python
now = datetime.now()
diff = now - then
diff
```

The difference between the datetime objects is represented as a `timedelta` object.

When it is converted to a string, it prints the difference in a convenient format (in terms of days, hours, minutes, seconds and microseconds):

```python
str(diff)
```

One can also access the days and seconds for the difference via attributes on the object:

```python
print(f"The difference is {diff.days} days and {diff.seconds} seconds")
```

Using a `datetime.timedelta` object, one can obtain dates certain number of years, months, or days in the past or in the future from a given one:

```python
from datetime import timedelta

diff = timedelta(days=1)
print(f"Now is {now}")
yesterday = now - diff
print(f"Yesterday was {yesterday}")
tomorrow = now + diff
print(f"Tomorrow will be {tomorrow}")
```

## Reading dates from a string

A common task when reading data from a file is to convert dates represented as strings into datetime objects. For that purpose, special symbols are used to interpret different words and digits that are commonly used to represent dates in text.

For example, suppose a date it represented as "2000-06-07 12:30:25". This string can be converted to a datetime object as follows:

```python
s = "2000-06-07 12:30:25"

# First special symbols are used to indicate that the input string should be read as:
# [year as four digits]-[month as two digits]-[day] [hours]:[minutes]:[seconds]
date_format = "%Y-%m-%d %H:%M:%S"

# convert from s to a datetime object using this format
d = datetime.strptime(s, date_format)

d
```

The formatting instructions are useful for reading data files, where dates can be represented in either the European or the North American convention. For example, in the European convention "06-07-2019" stands for "6th July 2019", while the same string in the North American convention stands for "7th June 2019". Thus it is very important to explicitly encode how dates in a data file should be read.
